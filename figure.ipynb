```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-1, 1, 51)
y = np.maximum(0, x)
fig, ax = plt.subplots(figsize=(3.8, 2.8)) # 限制图片大小
ax.plot(x,y)

plt.show()
```
